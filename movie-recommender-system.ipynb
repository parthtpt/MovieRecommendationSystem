```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action','Adventure','Fantasy','SciFi']
```

```python
#list of string into num
import ast
ast.literal_eval( '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
## conversion and change list of string into number

def conversion(object):
    L = []
    for i in ast.literal_eval(object):
        L.append(i['name'])
    return L    
```

```python
movies['genres'] = movies['genres'].apply(conversion)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(conversion)
```

```python
movies.head()
```

```python
#we need only three cast

def conversion3(object):
    L = []
    counter = 0
    for i in ast.literal_eval(object):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L    
```

```python
movies['cast'] = movies['cast'].apply(conversion3)
```

```python
movies.head()
```

```python
#only job in crew
def director(object):
    L = []
    for i in ast.literal_eval(object):
        if i['job'] == 'Director':
            L.append(i['name'])
            break       
    return L    
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
#changing string of overview in list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# removing space between words
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies_df = movies[['movie_id','title','tags']]
```

```python
movies_df
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:" ".join(x))
```

```python
movies_df.head()
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
movies_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)    
```

```python
movies_df['tags'] = movies_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(similarity[1],reverse = True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:5]
```

```python
def recommend(movie):
    movie_index = movies_df[movies_df['title'] == movie].index[0]
    distances =similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:5]
    
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
    return
```

```python
recommend('Batman Begins')
```

```python
recommend('The Avengers')
```
